```python
import pandas as pd
import numpy as np
from math import log
```

```python
exp_table = pd.read_table("../Expression/GSE157103_genes.tpm.tsv", sep="\t", index_col=0)
exp_table
```

```python
cov = exp_table[[name for name in exp_table.columns if name.startswith("C")]].T
non_cov = exp_table[[name for name in exp_table.columns if not name.startswith("C")]].T
```

```python
usa_data = pd.read_excel("../../../MO_COVID_CELL_PRESS/1-s2.0-S2405471220303719-mmc2.xlsx", sheet_name="A All pvalues", index_col=0)
usa_data = usa_data[usa_data["biomolecule_type"] == "transcript"]
usa_data
```

```python
for gene in usa_data["standardized_name"]:    
    cov_expression = cov[gene].mean()
    non_cov_expression = non_cov[gene].mean()
    
    fc = cov_expression / non_cov_expression
    fc = round(log(fc, 2), 3)
    
    expected = usa_data[usa_data["standardized_name"] == gene]["Log2(COVID-19/non-COVID-19)"].squeeze()
    expected = round(expected, 3)
    
    if not np.allclose(np.array([fc]), np.array([expected])):
        print(f"Gene: {gene}, Calculated: {fc}, Expected: {expected}")
```
